```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
sns.set()
```

```python
df = pd.read_csv('pokemon.csv', encoding='latin-1', index_col=0)
df.head()
```

```python
sns.lmplot(data=df, x='Attack', y='Defense')
```

```python
#tarefa 1
sns.lmplot(data=df, x='Attack', y='Defense', fit_reg=False, hue='Stage')
```

```python
sns.lmplot(data=df, x='Attack', y='Defense', fit_reg=False, hue='Stage')
plt.ylim(0,200)
plt.xlim(0,300)
```

```python
#tarefa 2
sns.lmplot(data=df, x='Attack', y='Defense', fit_reg=False, hue='Stage')
plt.ylim(0,200)
plt.xlim(0,300)
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('ataque e defesa')
```

```python
sns.boxplot(data=df)
```

```python
#tarefa 3
sns.boxplot(data=df.drop(columns=['Total','Stage','Legendary']))
```

```python
#Tarefa 4
sns.set_style('dark')
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(data=df, x= 'Type 1', y='Attack')
```

```python
pokemon_cores = [
    '#78C850',  # Grass
    '#F08030',  # Fire
    '#6890F0',  # Water
    '#A8B820',  # Bug
    '#A8A878',  # Normal
    '#A040A0',  # Poison
    '#F8D030',  # Electric
    '#E0C068',  # Ground
    '#EE99AC',  # Fairy
    '#C03028',  # Fighting
    '#F85888',  # Psychic
    '#B8A038',  # Rock
    '#705898',  # Ghost
    '#98D8D8',  # Ice
    '#7038F8'   # Dragon
]
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='Type 1', y='Attack', data=df, palette=pokemon_cores )
```

```python
#tarefa 5
sns.set_style('whitegrid')
plt.figure(figsize=(12,6))
sns.swarmplot(x='Type 1', y='Attack', data=df, palette=pokemon_cores)
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='Type 1', y= 'Attack', data=df, inner= None, palette= pokemon_cores)
sns.swarmplot(x='Type 1', y= 'Attack', data=df, color='k', alpha=0.7)
plt.title('Ataque por tipo de pokemon')
```

```python
#tarefa 6
df.drop(columns=['Total', 'Stage', 'Legendary']).corr()
```

```python
#tarefa 7
sns.histplot(data=df, x='Attack', kde=True )
```
