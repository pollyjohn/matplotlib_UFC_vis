```python
df['date']= pd.to_datetime(df['date'])
figths_by_year = df.groupby(df['date'].dt.year).agg({'R_fighter': 'count'})
```

```python
figths_by_year
```

```python
#tarefa 7
fig = plt.figure(figsize=(10,5))
fig.add_subplot(1,2,1)
plt.plot(figths_by_year.index, figths_by_year["R_fighter"])
plt.xlabel('Ano')
plt.ylabel('# lutas')
plt.title('lutas por ano')
```

```python
fights_by_class = df["weight_class"].value_counts()[:10]
```

```python
#tarefa 8 
figths_by_year.plot(kind='bar')
```

```python
fights_by_class.plot(kind='bar')
```

```python
fig = plt.figure(figsize=(20,5))
ax1 =  fig.add_subplot(1,2,1)
ax1.set_title('Lutas por peso')
fights_by_class.plot(kind='bar', ax=ax1)
ax2 = fig.add_subplot(1,2,2)
ax2.set_title('lutas por ano')
figths_by_year.plot(kind='bar', ax=ax2)
```
